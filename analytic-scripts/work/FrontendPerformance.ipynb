## Front end performance spike

### Related data

Data for this experiment is snapshotted at ./fe.csv which is source from a google spreadsheet

### Hypothesis

The frontend can naively fetch and render a large amount of data in a react-redux app

### Setup

* Created basic react-redux app that fetches product list
* Used json-server to generate large number of products
* Displayed products on webpage.


```r
df <- read.csv("./fe.csv", header = TRUE)
colnames(df) <- c("SizeOfRecordSet", "OverallLatency");
print(df)
```

```r
library(ggplot2)

g <- ggplot(df) + geom_point(aes(x = SizeOfRecordSet, y = OverallLatency)) 
g
```

## Conclusion

* At 40000+ records the UI becomes prohibitively non-performant (>10s)
